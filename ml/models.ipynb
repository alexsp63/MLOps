Модель глубокого обучения (ДОП)

```python
import torch
import torch.optim as optim
import torch.nn as nn

class NeuralNetwork(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, hidden_size3, hidden_size4, output_size):
        super(NeuralNetwork, self).__init__()
        self.l1 = nn.Linear(input_size, hidden_size1) 
        self.f1 = nn.Softmax()
        self.l2 = nn.Linear(hidden_size1, hidden_size2)  
        self.f2 = nn.Softmax()
        self.l3 = nn.Linear(hidden_size2, hidden_size3) 
        self.f3 = nn.Softmax()
        self.l4 = nn.Linear(hidden_size3, hidden_size4) 
        self.f4 = nn.Sigmoid()
        self.l5 = nn.Linear(hidden_size4, output_size) 
  
    def forward(self, inputs):
        return self.l5(self.f4(self.l4(self.f3(self.l3(self.f2(self.l2(self.f1(self.l1(inputs)))))))))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
scaler_y = MinMaxScaler().fit(y_train.reshape(-1, 1))
y_train_norm = scaler_y.transform(y_train.reshape(-1, 1))
scaler_x = MinMaxScaler().fit(X_train)
X_train_norm = scaler_x.transform(X_train)
X_test_norm = scaler_x.transform(X_test)

# обучение модели

network = NeuralNetwork(X.shape[1], 128, 256, 128, 64, 1)   # побольше нейронов
loss = nn.MSELoss() 
learning_rate = 0.1
X_train_norm = torch.from_numpy(X_train_norm).float()
y_train_norm = torch.from_numpy(y_train_norm).float()

optimizer = optim.Adam(network.parameters(), lr=learning_rate)

for epoch in range(100):
    # получение предсказаний
    y_pred = network(X_train_norm)

    # прогон через функцию потерь с предсказанными значениями
    data_loss = loss.forward(y_pred, y_train_norm)
    data_loss.backward()
    
    # обновление параметров сети
    optimizer.step()

    # обнуление градиента после обновления
    optimizer.zero_grad() 
    
    # if epoch !=0 and epoch % 20 == 0:
    print(f'epoch = {epoch}, loss = {data_loss}')
```

```python
predict_train_norm = network(X_train_norm).detach().numpy()
predict_test_norm = network(torch.from_numpy(X_test_norm).float()).detach().numpy()

predict_train = scaler_y.inverse_transform(predict_train_norm.reshape(-1, 1))
predict_test = scaler_y.inverse_transform(predict_test_norm.reshape(-1, 1))

calculate_metrics('My Neural Network', y_train, predict_train, y_test, predict_test)
```
